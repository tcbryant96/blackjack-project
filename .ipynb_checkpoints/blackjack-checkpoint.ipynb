```python
import random 
from IPython.display import clear_output
import time

class Blackjack:
    def __init__(self):
        self.player = []
        self.suits = ["♠", "♥", "♣", "♦","♠", "♥", "♣", "♦","♠", "♥", "♣", "♦","♠", "♥", "♣", "♦"]
        self.cards = ["Ace",2, 3, 4, 5, 6, 7, 8, 9, 10, "Jack", "Queen", "King","Ace",2, 3, 4, 5, 6, 7, 8, 9, 10, "Jack", "Queen", "King","Ace",2, 3, 4, 5, 6, 7, 8, 9, 10, "Jack", "Queen", "King","Ace",2, 3, 4, 5, 6, 7, 8, 9, 10, "Jack", "Queen", "King"]
        self.player_1 = []
        self.dealer = []
        self.player_t = 0
        self.dealer_t = 0
        self.player_ace_count = 0
        self.dealer_ace_count = 0
    def add_player(self):
        add = input("What is you name: ").title()
        self.player.append(add)
        clear_output()
        print(f"{add} has been added")
        play_game()
    def deal_cards(self):
        while len(self.player_1)< 2:
            suit = random.choice(self.suits)
            self.card = random.choice(self.cards)
            card_suit = (f"{self.card} of {suit}")
            self.player_1.append(card_suit)
            self.suits.remove(suit)
            self.cards.remove(self.card)
            if self.card == "Ace":
                self.player_ace_count += 1
            self.player_total()
           
        
        while len(self.dealer)< 2:
            suit = random.choice(self.suits)
            self.card = random.choice(self.cards)
            card_suit = (f"{self.card} of {suit}")
            self.dealer.append(card_suit)
            self.suits.remove(suit)
            self.cards.remove(self.card) 
            if self.card == "Ace":
                self.dealer_ace_count += 1
            self.dealer_total()
            
    def player_total(self):    
        face= ["King", "Queen", "Jack"]
        if self.card in face:
             self.player_t += 10
        if self.card == "Ace":
            self.player_t +=11
        if self.card in range(2,11):
            self.player_t += self.card
        if self.player_ace_count >1:
            self.player_t -= 10
            self.plaer_ace_count -= 1
        return self.player_t

    def dealer_total(self):
        face= ["King", "Queen", "Jack"]
        if self.card in face:
             self.dealer_t += 10
        if self.card == "Ace":
            self.dealer_t +=11
        if self.card in range(2,11):
            self.dealer_t += self.card
        if self.dealer_ace_count >1:
            self.dealer_t -= 10
            self.dealer_ace_count -= 1
        return self.dealer_t

       
            
    def start_game(self):
        clear_output()
        self.player_1 = []
        self.dealer = []
        self.player_t = 0
        self.dealer_t = 0
        self.suits = ["♠", "♥", "♣", "♦","♠", "♥", "♣", "♦","♠", "♥", "♣", "♦","♠", "♥", "♣", "♦"]
        self.cards = ["Ace",2, 3, 4, 5, 6, 7, 8, 9, 10, "Jack", "Queen", "King","Ace",2, 3, 4, 5, 6, 7, 8, 9, 10, "Jack", "Queen", "King","Ace",2, 3, 4, 5, 6, 7, 8, 9, 10, "Jack", "Queen", "King","Ace",2, 3, 4, 5, 6, 7, 8, 9, 10, "Jack", "Queen", "King"]
        self.dealer_ace_count = 0
        self.player_ace_count = 0
        print("Dealing cards...")
        time.sleep(2)
        self.deal_cards()
        print(self.player_1)
        amount = self.player_t
        if amount == 21:
            print(f"Oh My Lanta! BlackJack! You Win!!!!!")
            self.end_game()
        print (f"Your total is {amount}")
        dealer_cards = self.dealer[1]
        print(f"Dealers cards are:  X, {dealer_cards}")

        print("""What would you like to do?
                1. Stay
                2. Hit me""")
        action = input("Enter 1/ 2/ : ")
        if action == "1":
            self.stay()
        if action == "2":
            self.hit_me()

        
     
            
        
                    
        
       
    def new_card(self):
        suit = random.choice(self.suits)
        self.card = random.choice(self.cards)
        new_suit = (f"{self.card} of {suit}")
        self.player_1.append(new_suit)
        self.suits.remove(suit)
        self.cards.remove(self.card)
        face= ["King", "Queen", "Jack"]
        if self.card == "Ace":
            self.player_ace_count += 1
        if self.card in face:
            self.player_t += 10
        if self.card == "Ace":
            self.player_t +=11
        if self.card in range(2,11):
            self.player_t += self.card
        if self.player_ace_count >0 and self.player_t > 21:
            self.player_t -= 10
            self.player_ace_count -= 1

        return self.player_t
           
    def dealer_new_card(self):
        suit = random.choice(self.suits)
        self.card = random.choice(self.cards)
        new_suit = (f"{self.card} of {suit}")
        self.dealer.append(new_suit)
        self.suits.remove(suit)
        self.cards.remove(self.card)
        face= ["King", "Queen", "Jack"]
        if self.card == "Ace":
            self.dealer_ace_count += 1
        if self.card in face:
             self.dealer_t += 10
        if self.card == "Ace":
            self.dealer_t +=11
        if self.card in range(2,11):
            self.dealer_t += self.card
        if self.dealer_ace_count > 0 and self.dealer_t > 21:
            self.dealer_t -= 10
            self.dealer_ace_count -= 1
 
        return self.dealer_t
    
    def hit_me(self):
        self.new_card()
        clear_output()
        print("Dealing cards...")
        time.sleep(2)
        clear_output()
        print(f"Your cards are now: ")
        print(self.player_1)
        print(f"Your new total is {self.player_t}")
        if self.player_t == 21:
            print("You Hit 21, Nice Job, Dealers Turn")
            time.sleep(5)
            self.stay()
        if self.player_t > 21:
            print("You busted, you lose ( ◣‗◢)ψ")
            self.end_game()
        print("""What would you like to do?
                1. Stay
                2. Hit me""")
        action = input("Enter 1/ 2/ : ")
        if action == "1":
            self.stay()
        if action == "2":
            self.hit_me()
    def stay(self):
        clear_output()
        print("Dealer is now playing...")
        time.sleep(2)
        clear_output()
        print(self.dealer)
        print(f"Dealers total is {self.dealer_t}")
        print("Dealer is thinking...")
        time.sleep(2)
        if self.dealer_t == 21:
            self.compare()
        elif self.dealer_t > 21:
            print("The Dealer busted, you win!!!!")
            self.end_game()
        elif self.dealer_t < 17:
            self.dealer_new_card()
            self.stay()
        elif self.dealer_t < self.player_t:
            self.dealer_new_card()
            self.stay()
        else:
            self.compare()
    def compare(self):
        clear_output()
        print(f"Dealers cards are {self.dealer}, with a total of {self.dealer_t}")
        print(f"Your cards are {self.player_1}, with a total of {self.player_t}")
        target = 21
        x = target - self.player_t
        y= target - self.dealer_t
        if x < y:
            print("You win, Nice job!!!")
        if x == y:
            print("It's a tie...")
        if y != 0 and x == 0:
            print("Dealer has Blackjack, You lose ( ◣‗◢)ψ")
        if x > y:
            print("You lose ( ◣‗◢)ψ")
        
        self.end_game()
        
    def end_game(self):
        print("what would you like to do? 1.Play again 2. View rules 3. Main Menu 4. Quit")
        game = input("Enter 1/ 2/ : ")
        while game == "4":
            clear_output()
            print("Thanks for playing")
            break
        if game == "1":
            self.start_game()
        elif game == "2":
            self.view_rules()
        elif game == "3":
            play_game()
    def view_rules(self):
        clear_output()
        print("""***How To Play Blackjack, In 4 Easy Steps***
Yes, when it comes to blackjack strategy, there are loads of rules to learn. But for now, here’s your basic guide of how to play blackjack: 

Your goal in blackjack is to beat the dealer’s hand without going over 21. 
You’ll receive 2 cards at the beginning of each round, and you’ll add up the values of these cards. 
Cards 2-10 have face value; King, Queen, Jack are worth 10; and Aces are either a 1 or an 11 — it’s up to you to decide.
The dealer also draws two cards. The aim of the game is to beat his hand (have a higher hand) without going over 21.
If you would like the dealer to deal you another card, you tell him “hit”
If you do not want to be dealt another card, you “stand”
“Busting” is when the sum of your hand is over 21. 
The winner of the round is whoever has the highest hand without going bust. 
        """)
        play_game()
def play_game():
    play = Blackjack()
    print()
    print("""Welcome to BlackJack! 
            1. Add player
            2. View the rules
            3. Start playing
            4. Quit """)
    option = input("Enter number here: ")
    while option not in {'1', '2', '3', "4"}:
        option = input('Invalid option. Please enter 1-4')
    if option == '4':
        clear_output()
        print('Thanks for playing')
        exit()
    if option == "1":
        play.add_player()
    elif option == '2':
        play.view_rules()
    elif option == '3':
        play.start_game()



play_game() 
```
